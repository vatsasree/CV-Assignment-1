All the above process of converting from 3D coordinates to 2D coordinates with intermediate homogeneous coordinates' can be acheived/performed using this below function

```python
def project_3D_2D(P, X, n):
    X_homo_3D = np.concatenate((X, np.ones((n,1))), axis=1)
    X_homo_2D = np.matmul(P,X_homo_3D.T)
    X_2D = X_homo_2D.T[:,0:2]/X_homo_2D.T[:,2:]
    return X_2D
```

```python
def move_checkboard_by_n_positions(checkboard2, wireframe_2D, shifted_corners_2D):
    image = checkboard2
    
    original_points = np.float32(wireframe_2D)
    shifted_points = np.float32(shifted_corners_2D)

    mask = np.zeros_like(image[:, :, 0], dtype=np.uint8)
    cv2.fillPoly(mask, [original_points.astype(np.int32)], 255)
    M = cv2.getPerspectiveTransform(original_points, shifted_points)
    height, width = image.shape[:2]
    warped_mask = cv2.warpPerspective(mask, M, (width, height))
    warped_mask_inv = cv2.bitwise_not(warped_mask)
    warped_image = cv2.warpPerspective(image, M, (width, height))
    masked_region = cv2.bitwise_and(image, image, mask=warped_mask_inv)
    shifted_region = cv2.bitwise_and(warped_image, warped_image, mask=warped_mask)

    result = cv2.add(masked_region, shifted_region)

    for i in range(len(shifted_corners_2D)):
        start_point = (int(shifted_corners_2D[i][0]), int(shifted_corners_2D[i][1]))
        end_point = (int(shifted_corners_2D[(i + 1) % len(shifted_corners_2D)][0]), int(shifted_corners_2D[(i + 1) % len(shifted_corners_2D)][1]))
        cv2.line(result, start_point, end_point, (0,0,255), 15)

    for i in range(len(wireframe_2D)):
        start_point = (int(wireframe_2D[i][0]), int(wireframe_2D[i][1]))
        end_point = (int(wireframe_2D[(i + 1) % len(wireframe_2D)][0]), int(wireframe_2D[(i + 1) % len(wireframe_2D)][1]))
        cv2.line(result, start_point, end_point, (255,0,0), 15)

    plt.imshow(result)    
```
